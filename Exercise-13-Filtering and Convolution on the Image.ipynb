# <h2>Exercise: Blur the Face Using Manual Coordinates and Filters</h2>
<p>
Blur the face by manually entering the face coordinates and using the filters we read.
</p>
<p>
To do this, you need to:
</p>
<ul>
    <li>First, cut out the area related to the face (with numpy).</li>
    <li>Then blur the cut-out area of the face with large filters.</li>
    <li>Finally, place this cut-out blurred area in its original place with numpy.</li>
</ul>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'D:\exam\face_image.jpg')  # Replace with your image path

# Define manual coordinates for the face region (x, y, width, height)
# Example coordinates: adjust these based on your image
x, y, w, h = 150, 100, 200, 250

# Extract the face region using numpy slicing
face_region = image[y:y+h, x:x+w]

# Apply Gaussian blur to the face region with a large kernel for strong blur
blurred_face = cv2.GaussianBlur(face_region, (21, 21), 10)

# Place the blurred face back into the original image
image[y:y+h, x:x+w] = blurred_face

# Display the results
plt.figure(figsize=[10, 5])
plt.subplot(121); plt.imshow(cv2.cvtColor(cv2.imread(r'D:\exam\face_image.jpg'), cv2.COLOR_BGR2RGB)); plt.title("Original")
plt.subplot(122); plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)); plt.title("Face Blurred")
plt.show()

# Save the result
cv2.imwrite('blurred_face_image.jpg', image)
```
