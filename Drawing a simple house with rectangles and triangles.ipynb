```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from zipfile import ZipFile

# ساخت تصویر خالی
image = np.zeros((512, 512, 3), np.uint8)

# رسم بدنه خانه (مستطیل)
cv2.rectangle(image, (200, 250), (350, 400), (0, 255, 255), 3)  # زرد با ضخامت 3

# رسم سقف خانه (مثلث)
pts = np.array([[200, 250], [350, 250], [275, 150]], np.int32)  # سه نقطه برای مثلث
pts = pts.reshape((-1, 1, 2))
cv2.polylines(image, [pts], True, (0, 0, 255), 3)  # قرمز با ضخامت 3

# نمایش تصویر در ژوپیتر
plt.figure(figsize=(6, 6))  # تنظیم اندازه تصویر
plt.imshow(image[...,::-1])  # تبدیل BGR به RGB برای نمایش درست رنگ‌ها
#plt.axis('off')  # حذف محورها
plt.title("Simple House")  # عنوان تصویر
plt.show()

# ذخیره تصویر به صورت فایل PNG
cv2.imwrite("simple_house.png", image)

# ساخت فایل فشرده
with ZipFile("simple_house_project.zip", "w") as zipf:
    zipf.write("simple_house.png")

# بررسی حجم فایل
file_size = os.path.getsize("simple_house_project.zip") / (1024 * 1024)  # به مگابایت
if file_size < 10:
    print(f"فایل با موفقیت ساخته شد. حجم فایل: {file_size:.2f} MB")
else:
    print("حجم فایل بیشتر از 10 مگابایت است!")
```
