# Read the hidden text in the image by increasing the contrast:

(Because the color of the text is very slightly different from the background color, the text is not clear on the original image.)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر
image = cv2.imread(r'D:\exam\test\Exercise_12_Increasing Text Contrast\hidden.png')

# تبدیل به تصویر سطح خاکستری
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# اعمال بهبود کنتراست با استفاده از هیستوگرام برابر سازی
equalized = cv2.equalizeHist(gray)

# آستانه‌گذاری (Thresholding) برای برجسته کردن متن
_, thresh = cv2.threshold(equalized, 200, 255, cv2.THRESH_BINARY)

# نمایش تصاویر برای مقایسه
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Grayscale + Contrast Enhancement')
plt.imshow(equalized, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Final result (emphasis added)')
plt.imshow(thresh, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
