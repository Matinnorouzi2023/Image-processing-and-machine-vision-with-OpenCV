```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  
image = cv2.imread('D:/exam/face_image.jpg')
if image is None:
    raise FileNotFoundError("Image file not found. Please provide a valid image in 'images/sample.jpg'.")

#     RGB     Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#  
(h, w) = image.shape[:2]
center = (w // 2, h // 2)  #    

# 1.   ( 45 )
angle = 45  #  
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1.0)  #  
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (w, h))

# 2.   (10   )
translation_matrix = np.float32([[1, 0, 10], [0, 1, 0]])  #  
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (w, h))

#  
plt.figure(figsize=[15, 5])

plt.subplot(131)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(rotated_image)
plt.title(f'Rotated Image ({angle} degrees)')
plt.axis('off')

plt.subplot(133)
plt.imshow(translated_image)
plt.title('Translated Image (10 units right)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
