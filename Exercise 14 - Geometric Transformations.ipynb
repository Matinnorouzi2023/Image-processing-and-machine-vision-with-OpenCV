```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری تصویر
image = cv2.imread('D:/exam/face_image.jpg')
if image is None:
    raise FileNotFoundError("Image file not found. Please provide a valid image in 'images/sample.jpg'.")

# تبدیل تصویر به فرمت RGB برای نمایش صحیح در Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ابعاد تصویر
(h, w) = image.shape[:2]
center = (w // 2, h // 2)  # مرکز تصویر برای چرخش

# 1. چرخش تصویر (مثلاً 45 درجه)
angle = 45  # زاویه چرخش
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1.0)  # ماتریس چرخش
rotated_image = cv2.warpAffine(image_rgb, rotation_matrix, (w, h))

# 2. انتقال تصویر (10 واحد به راست)
translation_matrix = np.float32([[1, 0, 10], [0, 1, 0]])  # ماتریس انتقال
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (w, h))

# نمایش تصاویر
plt.figure(figsize=[15, 5])

plt.subplot(131)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(rotated_image)
plt.title(f'Rotated Image ({angle} degrees)')
plt.axis('off')

plt.subplot(133)
plt.imshow(translated_image)
plt.title('Translated Image (10 units right)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
